# Agrupamiento de lenguajes con redes de co-ocurrencia de palabras 

## Hipótesis

Las topología de las redes de co-ocurrencia de palabras reflejan las relaciones genealógicas entre los lenguajes.

## Predicción

Un análisis de agrupamiento por similitud de algunas medidas topológicas de las redes de co-ocurrencia de un mismo texto en distintos lenguajes recuperará las relaciones genealógicas entre éstos

## Método

### Definición de las redes

Se definieron 9 gráficas ($G(V,E)$) a partir del la obra *Manifiesto del partido comunista* de Karl Marx. Cada red corresponde al mismo texto en los siguientes lenguajes: español, inglés, portugués, italiano, francés, alemán, turco, ruso y vasco. En estas redes cada vértice ($v_i$) representan una palabra distinta y existe un enlcace ($e_i$) entre la palabra $v_n$ y $v_m$ si $v_n$ es vecina a $v_m$. Por ejemplo, para las oraciones "El perro está enojado con el gato. El niño está llorando.", se obtienen el siguiente conjunto de vérices ($V$) y de enlaces ($E$):

$V$ = *{el,perro,está,enojado,con,gato,.,niño,llorando}*

$E$ = *{{el,perro},{perro,está},{está,enojado},{enojado,con},{con, el}, {el,gato},{gato,.},{.,el},{el,niño},{niño,llorando},{llorando,.}}*


```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import re
import random
import pandas as pd
```

```python
## se leen los archivos
idiomas = ['aleman','arabe','catalan','checo','chino','coreano','danes','espanol','euskera','finlandes','frances','griego','holandes','hungaro','indonesio','ingles','italiano','macedonio','noruego','persico','polaco','portugues','rumano','ruso','serbio','slovaco','sloveno','sueco','tailandes','turco','vietnamita']
#idiomas = ['aleman','catalan','espanol','frances','ingles','italiano','portugues']
textos = []
for idioma in idiomas:
    with open('./archivos/manifiesto/manifiesto-' + idioma + '.txt') as archivo:
        textos.append( archivo.read() )
```

```python
textos[30]
```

```python
## se limpian los textos
def limpiarTexto (text):
    text = text.lower()
    skips = ['...','…','’',',','„',';',':','"','“',"'",'¡','!','¿','?','\n','-','–','—','”','»','«','(',')','‘','\t\u3000\u3000','›','‹','1','2','3','4','5','6','7','8','9','0','^','ᵉ','\\ ','[','*',']','\x0c','..','’’','°']
    for i in skips:
        text = text.replace(i, ' ')
        text = re.sub(' +', ' ', text)
    return text

textos_limpios = []
for texto in textos:
    textos_limpios.append( limpiarTexto(texto) )
```

```python
## se definen los enlaces de las redes
def crearEnlaces( texto ):
    texto_lista = re.findall(r"[\w']+|[.,!?;]", texto) 
    pares_palabras = []
    for i in range( len(texto_lista) - 1):
        par = [ texto_lista[i], texto_lista[i + 1]]
        pares_palabras.append(par)
    return pares_palabras

listas_enlaces = []
for texto_limpio in textos_limpios:
    listas_enlaces.append( crearEnlaces(texto_limpio) )
```

```python
## se construyen las redes en un diccionario
DGs = {}
for lista_enlaces, idioma in zip( listas_enlaces , idiomas ):
    DGs.update( { idioma : nx.Graph(lista_enlaces) } )
```

```python
## se revisa si son gráficas es conectada o fuertemente conectadas
if nx.is_directed(DGs.get('espanol')):
    print('¿Fuertemente conectada? \n' )
    for key in DGs:
        print(key + ':' + str(nx.is_strongly_connected(DGs.get(key))))
else:
    print('¿Conectada? \n' )
    for key in DGs:
        print(key + ':' + str(nx.is_connected(DGs.get(key))))
```

```python
## se grafica un ejemplo reducido de la red construida
numero_de_nodos = 50
GReducidaEjemplo = nx.Graph(listas_enlaces[idiomas.index('espanol')][:numero_de_nodos - 1])

figura = plt.figure(figsize=(11,10))
  
pos = nx.kamada_kawai_layout(GReducidaEjemplo)

nx.draw(GReducidaEjemplo,
        pos,
        node_color='w',
        edgecolors = 'grey',
        linewidths=3,
        node_size=100,
        node_shape='o',
        width=3,
        edge_color='grey')

m = 0.04
for key, value in pos.items():
    pos[key] = np.array( [(pos[key][0]), (pos[key][1]+m)] ) 
    
labels = nx.draw_networkx_labels(GReducidaEjemplo,
                                 pos,
                                 font_size = 18,
                                 alpha=0.9,)
figura.savefig('red-paramo-2.png')
```

### Análisis cuantitativo de las redes

Se caluclaron las siguentes medidas para cada red:
1. número de nodos ($V$)
- número de enlaces ($E$)
- grado promedio ($k_{prom}$)
- centralidad de grado promedio ($C_{grado}$)
- centralidad de eigenvector promedio ($C_{eig}$)
- centralidad de Katz promedio promedio ($C_{katz}$)
- centralidad de intermediación promedio ($C_{bet}$)
- centralidad de cercanía promedio ($C_{clos}$)
- longitud promedio ($long_{prom}$)
- coeficiente de clustering ($clus$)
- diámetro ($Diam$)
- ajuste a la ley de potenia ($alpha$)
- centralización de la red ($centralización$)
- heterogeneidad de la red ($hetereogenedidad$)

```python
from scipy import stats

if nx.is_directed(DGs.get('espanol')):
    j = 1
else:
    j = 2

idioma = []
V = []
E = []
k_prom = []
#C_grado = []
#C_eig = []
#C_katz = []
#C_bet = []
#C_clos = []
long_prom = []
clus = []
diam = []
alpha = []
centralizacion = []
heterogeneidad = []

for key in DGs:
    idioma.append(key)
    V.append( len(DGs.get(key).nodes()) )
    E.append( len(DGs.get(key).edges()) )
    k_prom.append( j * len( DGs.get(key).edges() ) / len( DGs.get(key).nodes() ) )
    #C_grado.append( np.mean( list( nx.degree_centrality(DGs.get(key)).values() ))  )
    #C_eig.append( np.mean( list( nx.eigenvector_centrality( DGs.get(key) ).values() ) ) )
    #C_katz.append( np.mean( list( nx.katz_centrality_numpy( DGs.get(key)).values() ) ) )
    #C_bet.append( np.mean( list( nx.betweenness_centrality( DGs.get(key) ).values() ) ) )
    #C_clos.append( np.mean( list( nx.closeness_centrality( DGs.get(key) ).values() ) ) )
    long_prom.append( nx.average_shortest_path_length(DGs.get(key)) )
    clus.append( nx.average_clustering( DGs.get(key) ) )
    diam.append( nx.diameter( DGs.get(key) ) )
    alpha.append( 1 + len(np.array( list( dict( DGs.get(key).degree ).values() ) ))*(sum( np.log(np.array( list( dict( DGs.get(key).degree ).values() ) )/ min(np.array( list( dict( DGs.get(key).degree ).values() ) ))) ))**(-1) )
    centralizacion.append( ( max( np.array( list( dict( DGs.get(key).degree ).values() ) ) ) / len( DGs.get(key).nodes ) ) - ( ( j * len( DGs.get(key).edges() ) / len( DGs.get(key).nodes() ) ) / ( len( DGs.get(key).nodes() ) - 1 ) ) )
    heterogeneidad.append( ( np.var( np.array( list( dict( DGs.get(key).degree ).values() ) ) ) )**(1/2) / ( j * len( DGs.get(key).edges() ) / len( DGs.get(key).nodes() ) )  )
    
```

```python
ajuste_dist_grado_R_sqrt = []
ajuste_dist_grado =  []
ajuste_dist_nadc_R_sqrt = []
ajuste_dist_nadc = []

for key in DGs:
    grados = np.array( list(dict( DGs.get(key).degree ).values() ) )
    x, y = np.unique(grados, return_counts = True)
    y = np.log10(y)
    x = np.log10(x)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    ajuste_dist_grado.append( -slope )
    ajuste_dist_grado_R_sqrt.append( r_value**2 )
    
for key in DGs:
    x = list( nx.average_degree_connectivity( DGs.get(key) ).keys() )
    y = list( nx.average_degree_connectivity( DGs.get(key) ).values() )
    slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(x), np.log(y))
    ajuste_dist_nadc.append( -slope )
    ajuste_dist_nadc_R_sqrt.append( r_value**2 )
    
```

```python
def graficarEscalaLineal(G, lab):
    grados = np.array( list(dict( G.degree ).values()))
    y,x = np.histogram(grados , bins = max(grados))
    x_med = 0.5*(x[:-1] + x [1:])
    plt.plot(x_med, (y/len( G.nodes())), '.', label = lab)
    plt.xlabel('$k$', fontsize= 18)
    plt.ylabel('$p_{k}$', fontsize= 18)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

def graficarBinLinear(G, lab):
    grados = np.array( list(dict( G.degree ).values()))
    y,x = np.histogram(grados, bins = max(grados) )
    x_med = 0.5*(x[:-1] + x [1:])
    plt.plot(x_med, (y/len( G.nodes())), '.', label = lab)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('$k$', fontsize= 18)
    plt.ylabel('$p_{k}$', fontsize= 18)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    
def graficarBinLog(G, lab):
    grados = np.array( list(dict( G.degree ).values()))
    log_bin = np.logspace(0, np.log10(max(grados)), 25)
    y,x = np.histogram(grados, bins = log_bin )
    x_med = 0.5*(x[:-1] + x [1:])
    plt.plot(x_med, (y/len( G.nodes())), '.', label = lab)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('$k$', fontsize= 18)
    plt.ylabel('$p_{k}$', fontsize= 18)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
def graficarAcum(G, lab):
    grados = np.array( list(dict( G.degree ).values()))
    y,x = np.histogram(grados, bins = max(grados) )
    y_cum = np.flip( np.cumsum( np.flip( y ) ) )
    x_med = 0.5*(x[:-1] + x [1:])
    plt.plot(x_med, (y_cum/len( G.nodes())), '.', label = lab)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('$k$', fontsize= 18)
    plt.ylabel('$P_{k}$', fontsize= 18)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

def graficarBinLogYAjuste(G, lab, alfa, col):
    grados = np.array( list(dict( G.degree ).values()))
    log_bin = np.logspace(0, np.log10(max(grados)), 25)
    y,x = np.histogram(grados, bins = max(grados) )
    x_med = 0.5*(x[:-1] + x [1:])
    plt.plot(x_med, (y/len( G.nodes())), '.', color = col, alpha=0.1)
    plt.plot(x_med, 1*x_med**(-alfa), color = col, label = lab)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('$k$', fontsize= 18)
    plt.ylabel('$p_{k}$', fontsize= 18)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
